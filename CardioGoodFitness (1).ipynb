```python
#Relationship between two categorical variables
pd.crosstab(data['Product'],data['Gender'])
```

```python
# Calculate the counts of each gender
gender_counts = data['Gender'].value_counts()
gender_counts
```

```python
# Plotting the pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Gym Goers by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Calculate the counts of each gender
gender_counts = data['Gender'].value_counts()
# Plotting the donut chart
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title('Proportion of Gym Goers by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# Display the plot
plt.show()
```

```python
product_counts = data['Product'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(product_counts.index, product_counts.values, color=['blue', 'green', 'orange'])
plt.xlabel('Product')
plt.ylabel('Count')
plt.title('Number of Participants by Product')
plt.show()
```

```python
#Relationship between two categorical variables
sns.countplot(x="Gender", hue = "Product", data = data)
```

```python
crosstab = pd.crosstab(data['Product'], data['Gender'])
crosstab.plot(kind='bar', stacked=True, figsize=(8, 6), color=['pink', 'blue'])
plt.xlabel('Product')
plt.ylabel('Count')
plt.title('Relationship between Product and Gender')
plt.legend(title='Gender')
plt.show()
```
