```python
#treating using log
import numpy as np
for col in df.columns:
    if df.skew().loc[col]>0.55:
        df[col]=np.log1p(df[col])
```

```python
x=df.drop(columns='Sales ($)')
y=df['Sales ($)']
print(x)
print(y)
```

```python
#Dimension and size of x
print(x.ndim)
print(x.size)
```

```python
#Dimension and size of x
print(y.ndim)
print(y.size)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=20)
```

```python
x_train,x_test,y_train,y_test
```

# Linear Regression ModelÂ¶

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_test=lr.predict(x_test)
y_pred_test
```

```python
#compare y_pred_test with y_test(actual o/p)
from sklearn import metrics 
print("Mean Absolute Error: ",metrics.mean_absolute_error(y_test,y_pred_test))
print("Mean Squared Error: ",metrics.mean_squared_error(y_test,y_pred_test))
print("Root Mean Squared Error: ",metrics.mean_squared_error(y_test,y_pred_test,squared=False))
print("R2 Score: ",metrics.r2_score(y_test,y_pred_test))
```

```python
#Scatter plot to depict Actual vs Predicted Values
plt.scatter(y_test,y_pred_test)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title('Actual vs Predicted Values using Test Data')
plt.show()
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
df_new
```

```python
x=df_new.drop(columns='Sales ($)')
y=df_new['Sales ($)']
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_test=lr.predict(x_test)
y_pred_test
```

```python
#compare y_pred_test with y_test(actual o/p)
from sklearn import metrics 
print("Mean Absolute Error: ",metrics.mean_absolute_error(y_test,y_pred_test))
print("Mean Squared Error: ",metrics.mean_squared_error(y_test,y_pred_test))
print("Root Mean Squared Error: ",metrics.mean_squared_error(y_test,y_pred_test,squared=False))
print("R2 Score: ",metrics.r2_score(y_test,y_pred_test))
```
